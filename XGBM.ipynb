```python
from numpy import loadtxt
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
dataset = loadtxt('/content/pima-indians-diabetes.data.csv', delimiter=",")
X = dataset[:,0:8]
Y = dataset[:,8]
```

```python
seed = 7
text_size = 0.33
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=text_size, random_state=seed)
```

```python
model = XGBClassifier()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
predictions = [round(value) for value in y_pred]
```

```python
accuracy = accuracy_score(y_test,predictions)
```

```python
print("Accuracy: %.2f%%"% (accuracy * 100.0))
```

```python
data = loadtxt('/content/pima-indians-diabetes.data.csv',delimiter=",")
```

```python
X = dataset[:,0:8]
```

```python
Y = dataset[:,8]
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.25, random_state = 0)
```

```python
import lightgbm as lgb
d_train = lgb.Dataset(x_train, label=y_train)
```

```python
params = {}
params['learn_rate'] = 0.03
params['boosting_type'] = 'gdbt'
params['objective'] = 'binary'
params['metric'] = 'binary_logloss'
params['sub_feature'] = 0.5
params['num_leaves'] = 10
params['min_data'] = 50
params['max_depth'] = 10
```

```python
clf = lgb.train(params, d_train, 100)
```
